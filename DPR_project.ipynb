프로젝트 : 강의평 분석 자동화 프로젝트

강의평은 학생들이 수강 신청을 할 때 중요한 참고 자료가 된다. 긍정적이거나 부정적인 강의평은 어떤 강의를 선택할지 결정하는 데 큰 영향을 미칠 수 있다.교수진과 대학 행정팀은 강의평을 통해 강의의 장단점을 파악하고, 강의 내용을 개선하거나 보완할 수 있다. 이를 통해 교육의 질을 향상시킬 수 있다.
강의평은 길이, 내용, 표현 방식이 매우 다양하다. 따라서 강의평을 분석하고 이해하는 것은 쉽지 않다. 강의평은 단순히 긍정적 또는 부정적으로 분류될 뿐만 아니라, 그 안에 담긴 감정이나 의견을 정확하게 이해하는 것이 중요하다. 예를 들어, 혼합된 감정(긍정적이면서 부정적인 요소가 공존하는 강의평)을 분석하는 것은 특히 중요히다.
수작업으로 강의평을 분석하는 것은 시간과 비용이 많이 드는 작업이다. 자동화된 강의평 분석 시스템을 구축하면 이 과정을 크게 효율화할 수 있다. 또한 자동화된 시스템은 실시간으로 강의평을 분석하고, 즉각적인 피드백을 제공할 수 있다. 이는 교수진이 신속하게 대응하고, 필요한 조치를 즉시 취할 수 있게 한다.

1. 프로젝트 목표 : 모델을 활용하여 강의평 텍스트의 감정을 정확하게 분석하고, 이를 통해 학생과 교수진 모두에게 유용한 정보를 제공하는 것을 목표로 한다. 데이터 증강 기법을 활용하여 모델의 학습 데이터를 풍부하게 하고, 이를 통해 분석 모델의 성능을 향상시키는 것을 목표로 한다.
결론적으로 이 프로젝트는 자연어 처리(NLP) 기술과 데이터 증강 기법을 활용하여 강의평 텍스트를 보다 정확하게 분석하고, 이를 통해 학생과 교수진 모두에게 가치 있는 인사이트를 제공하는 것을 목표로 한다. 강의평 분석의 자동화는 효율성을 높이고, 실시간 피드백을 가능하게 하며, 궁극적으로는 더 나은 교육의 질을 관리하고 향상시키는 데 기여할 것이다.

2. 데이터 패턴 종류 : 프로그램에서 다루는 데이터는 텍스트 형식의 강의 평가와 각 평가에 대한 감정 레이블(긍정적, 부정적, 혼합)이다. 텍스트 데이터는 비구조화된 데이터로, 자연어 처리 기법을 통해 분석되고 전처리된다. 감정 레이블은 정수 인코딩과 원-핫 인코딩을 통해 모델에 입력된다.

3. 데이터 수집 과정 및 augmentation 사용 기술 :

 데이터는 공개된 강의 평가 데이터를 수집하여 구성된다. 이 데이터셋은 강의 평가(review)와 각 평가에 대한 감정(sentiment) 레이블로 구성되어 있다. 데이터를 CSV 파일 형식으로 로드하여 Pandas 데이터프레임으로 처리한다.

 데이터 증강(Data Augmentation)은 데이터셋의 다양성을 높이고 모델의 일반화 능력을 향상시키기 위해 사용되는데, 이 프로젝트에서는 nlpaug 라이브러리를 사용하여 다음과 같은 텍스트 증강 기법을 적용했다:
 ContextualWordEmbsAug: 문맥적 단어 임베딩을 사용하여 단어를 대체한다.
RandomWordAug (swap): 임의의 단어를 다른 위치로 교환한다.
RandomWordAug (delete): 임의의 단어를 삭제한다.
증강 기법을 통해 데이터셋의 다양성을 증가시켜 모델의 일반화 능력을 향상시킨다. 특히 지금처럼 데이터셋이 작은 경우 유용하다. 각 리뷰 텍스트에 대해 세 가지 증강 기법을 적용하여 세 개의 새로운 텍스트를 생성한다. 원본 텍스트도 포함하여 총 네 개의 텍스트를 만든다. 증강된 텍스트와 레이블을 리스트에 추가한다. 데이터 증강을 통해 데이터의 양을 늘리고, 모델의 학습에 도움이 되는 다양한 입력을 제공하였다.

4. 전이학습 및 타인의 데이터는 사용하지 않았다.

5. 데이터셋의 크기와 개수는 딥러닝 모델을 훈련시키기에 충분해야 한다. 일반적으로 딥러닝 모델은 많은 양의 데이터를 필요로 하므로 데이터 증강 기법을 통해 데이터를 늘리는 것이 중요하다.
원본 데이터 개수: 데이터에서는 133개의 리뷰가 포함되어 있다.
증강 데이터 개수: 증강 기법을 통해 데이터셋의 크기가 원본의 4배로 증가한다.
이 정도 크기의 데이터셋은 딥러닝 모델을 훈련시키기에 적절하며, 학습 시간과 성능 사이의 균형을 맞출 수 있다.

6. 데이터 전처리 및 패딩 : 텍스트 데이터를 정수 시퀀스로 변환하고, 모델 입력 크기를 맞추기 위해 패딩한다. Tokenizer를 사용하여 텍스트를 토큰화하고 숫자 시퀀스로 변환한다.pad_sequences를 사용하여 모든 시퀀스의 길이를 maxlen으로 맞춘다. 텍스트 데이터를 숫자 시퀀스로 변환한 후, 패딩을 통해 길이를 동일하게 맞춘다. 이는 모델이 입력으로 일정한 길이의 데이터를 필요로 하기 때문이다.
레이블 인코딩 : 감정 레이블을 숫자로 인코딩한 후, 원-핫 인코딩을 적용한다. 모델의 출력이 3개의 클래스(positive, negative, mixed)에 대한 확률 값을 예측하므로, 레이블을 원-핫 인코딩하여 모델의 출력과 비교할 수 있다.
딥러닝 구조 : 이 모델은 총 3개의 층으로 구성되어 있는데 Embedding 레이어를 사용하여 단어를 임베딩 벡터로 변환한다. 또, 양방향 LSTM을 사용하여 시퀀스 데이터를 처리하며, 드롭아웃 레이어를 통해 과적합을 방지한다. 최종 출력 레이어는 소프트맥스 활성화 함수를 사용하여 세 가지 감정 클래스(긍정적, 부정적, 혼합)를 예측한다. 이 모델은 텍스트 데이터를 효과적으로 처리하고 감정 레이블을 예측할 수 있도록 설계되었다.

7. 결과 분석 및 설명 : 모델을 훈련한 후, 테스트 데이터로 평가한 결과, 정확도는 약 86%로 나타났다. 이는 주어진 데이터에 대해 상당히 높은 정확도를 보여준다.
오차행렬과 각 클래스에 대한 정밀도(Precision), 재현율(Recall), F1 점수(F1-score) 및 지원(Support)을 보여주는 결과보고서, PR 곡선, AUC를 통해 분석해보았다. PR 곡선 아래의 면적인 AUC는 모델의 성능을 나타내며, AUC가 1에 가까울수록 모델의 성능이 좋다. 이 모델의 AUC는 0.80으로, 양호한 성능을 보인다. 또한 오차행렬을 통해 결과적으로 이 모델은 positive 및 negative 클래스에 대해 높은 성능을 보이고 있으나, mixed 클래스에 대해서는 성능이 낮은 것으로 나타났다. 이는 mixed 클래스의 샘플 수가 적고, 데이터 불균형으로 인한 영향일 수 있다.

```python
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
import nlpaug.augmenter.word as naw
from tensorflow.keras import models, layers

# 깃허블르 통한 데이터 로드
url = 'https://raw.githubusercontent.com/senness39/datapatternhw/main/lecture_reviews.csv'
df = pd.read_csv(url)

texts = df['review'].tolist()
labels = df['sentiment'].tolist()

# 데이터 증강 함수
def augment_text(text, augmenters, num_aug=1):
    augmented_texts = []
    for _ in range(num_aug):
        augmented_text = text
        for augmenter in augmenters:
            augmented_text = augmenter.augment(augmented_text)
        augmented_texts.append(augmented_text)
    return augmented_texts

augmenters = [
    naw.ContextualWordEmbsAug(model_path='bert-base-uncased', action="substitute"),
    naw.RandomWordAug(action="swap"),
    naw.RandomWordAug(action="delete"),
]

# 데이터 증강
augmented_texts = []
augmented_labels = []
for text, label in zip(texts, labels):
    augmented_texts.extend(augment_text(text, augmenters, num_aug=3))
    augmented_texts.append(text)  # 증강되지 않은 원본 데이터를 추가합니다.
    augmented_labels.extend([label] * 4)  # 원본 데이터를 제외하고 추가된 데이터의 개수는 4개입니다.

# 텍스트 전처리 및 패딩
tokenizer = Tokenizer(num_words=10000, oov_token='<OOV>')
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
maxlen = 100
data = pad_sequences(sequences, maxlen=maxlen, truncating='post', padding='post')

# 레이블 인코딩
label_dict = {'positive': 0, 'negative': 1, 'mixed': 2}
labels = [label_dict[label] for label in labels]
labels = to_categorical(labels, num_classes=3)

# 모델 생성
model = models.Sequential([
    layers.Embedding(10000, 128, input_length=maxlen),
    layers.Bidirectional(layers.LSTM(128, return_sequences=True)),
    layers.Bidirectional(layers.LSTM(128)),
    layers.Dropout(0.5),
    layers.Dense(64, activation='relu'),
    layers.Dense(3, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(data, labels, epochs=10, batch_size=32, validation_split=0.2)

# 테스트 데이터 평가
loss, accuracy = model.evaluate(data, labels)
print(f"테스트 정확도: {accuracy}")
```

```python
# 결과 분석
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

predictions = np.argmax(model.predict(data), axis=-1)

true_labels = np.argmax(labels, axis=-1)

# 오차 행렬
conf_matrix = confusion_matrix(true_labels, predictions)
print("오차 행렬:")
print(conf_matrix)

# 분류 보고서
class_names = ['positive', 'negative', 'mixed']
print("\n분류 보고서:")
print(classification_report(true_labels, predictions, target_names=class_names))
```

```python
from sklearn.metrics import precision_recall_curve, auc
import matplotlib.pyplot as plt

predictions = model.predict(data)

precision, recall, thresholds = precision_recall_curve(labels[:, 1], predictions[:, 1])

# Pr그래프
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, color='blue', lw=2, label='PR Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='lower left')
plt.show()

auc_score = auc(recall, precision)
print(f'AUC: {auc_score}')
```
